```python
from models.posts import Posts
from utils.get_semantic import get_embedding
from models.semantic_embeddings import SemanticEmbeddings
```

```python
from app import app
with app.app_context():
    posts = Posts.query.filter(Posts.title!=None).limit(5).offset(0).all()
```

```python
titles = [p.title  for p in posts]
titles = get_embedding(titles)
```

```python
print([t.shape for t in titles])
```

```python
from models import db
with app.app_context():
    inputs = []
    for index, p in enumerate(posts):
        indi = SemanticEmbeddings(fieldid=p.fieldid, postid=p.id, embedding=titles[index])
        inputs.append(indi)
```

```python
inputs
```

```python
with app.app_context():
    with db.session() as session:
        session.add_all(inputs)
        session.commit()
```
